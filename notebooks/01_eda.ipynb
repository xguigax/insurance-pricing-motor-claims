```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/freq_data.csv")

# Basic info
df.shape, df.head()
```

```python
df.describe(include="all")
```

```python
sns.countplot(x=df['ClaimNb'])
plt.title("Number of Claims Distribution")
plt.show()
```

```python
sns.histplot(df['DrivAge'], bins=40, kde=True)
plt.title("Driver Age Distribution")
plt.show()
```

```python
sns.histplot(df['Exposure'], bins=40, kde=True)
plt.title("Exposure Distribution")
plt.show()
```

```python
sns.countplot(y=df['Region'], order=df['Region'].value_counts().index)
plt.title("Region Distribution")
plt.show()
```
